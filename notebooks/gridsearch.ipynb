# Results on grid searching hyperparameters (optimization costs, etc.)

```python
def experiment_df(name, path='/data/manifolds/pybullet_eval/'):
    dfs = []
    all_params = []
    for i, trial in enumerate(os.listdir(path + name)):
        trial_key = trial.split('GF_')[0]
        trial_params = [float(x[2:]) for x in trial_key.split('_') if x != '']

        df = pd.read_csv(f'{path}/{name}/{trial}/metrics.csv')
        df["execution"] = df["execution"].astype(float)
        df["planning"] = df["planning"].astype(float)
        df = df.drop('scene_idx', axis=1)
        if df["execution"].sum() >= 2:
            print(trial)
            print(df)
        for j in range(len(df)):
            all_params.append(trial_params)
        dfs.append(df)
        
    all_dfs = pd.concat(dfs)
    all_params = pd.DataFrame(all_params)
    df = pd.concat([all_dfs.reset_index(), all_params], axis=1)

    names = {0: 'Smth',
            1: 'Obs',
            2: 'Grasp',
            3: 'Step',
            4: 'Num',
            5: 'Thresh',
           }
    df.rename(columns=names,
              inplace=True)
    return df

def display_means(df):
    print("Smoothness Weight")
    display(df.groupby('Smth').mean())
    print("Obstacle Weight")
    display(df.groupby('Obs').mean())
    print("Grasp Weight")
    display(df.groupby('Grasp').mean())
    print("Step Size")
    display(df.groupby('Step').mean())
    print("Num Steps")
    display(df.groupby('Num').mean())
    print("Goal Thresh")
    display(df.groupby('Thresh').mean())
```
